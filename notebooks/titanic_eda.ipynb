```python
import pandas as pd
#Read the data provided.
data=pd.read_csv('../data/titanic.csv')
print(data.head())
#No missing columns or rows no need for further editing
#used to see the parameters provided.
print(data.keys())
#Lets visualize the raw data
print(data.hist(figsize=(30, 14)))
#Note this only shows numerical data because it cant directly reference categorical data for that you would need to use replace function and assign numerical values to the data
```

```python
#describes the procured data into founding commanility within thier own data set.
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Survival rate by Sex
plt.figure(figsize=(8,5))
sns.barplot(data=data, x='Sex', y='Survived')
plt.title("Survival Rate by Sex")
plt.show()

# Survival rate by Class
plt.figure(figsize=(8,5))
sns.barplot(data=data, x='Pclass', y='Survived')
plt.title("Survival Rate by Class (Pclass)")
plt.show()
```
